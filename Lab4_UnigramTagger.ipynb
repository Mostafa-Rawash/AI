```python
import nltk
nltk.download("brown")
# from nltk.corpus import brown
from nltk.tag import UnigramTagger
```

```python
# to get all dataset sentence
test_sent = brown.sents()
# for i in test_sent:
#   print(i)
```

```python
unigram_tagger = UnigramTagger(brown.tagged_sents())
# to remove duplication in the dataset and make it in one array 
words = []
for  i in test_sent:
  for j in i:
    if j not in words:
      words.append(j)
```

```python
words
```

```python
taged  = []
for tok, tag in unigram_tagger.tag(words):
    taged.append((tok, tag))

```
